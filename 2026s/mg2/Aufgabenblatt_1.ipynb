### Aufgabe 1

Differenzieren Sie folgende rationale Funktion:

$$f(x)=\frac{1-x^2}{1+x^2}$$

#### Using the Quotient Rule
$$\boxed{\left(\frac{u}{v}\right)' = \frac{u'v - uv'}{v^2}}$$
##### Find Sub-Derivatives
Substitute $u = (1-x^2)$, and $v = (1+x^2)$, then using the power rule:

$\frac{d}{dx}(1-x^2) = -2x$, similarly $\frac{d}{dx}(1+x^2) = 2x$. 
#### Plug-in Derivatives
$$
 f'(x)=\frac{( -2x)(1+x^{2})-(1-x^{2})(2x)}{(1+x^{2})^{2}}
      =\frac{-2x-2x^{3}-2x+2x^{3}}{(1+x^{2})^{2}}
      =\frac{-4x}{(1+x^{2})^{2}}
$$

#### Check Calculations

```python
derivative = diff((1 - x**2) / (1 + x**2), x)
derivative.simplify()
```

#### Therefore, the Solution Is :
$$\boxed{f'(x) = -\frac{4x}{(1+x^2)^2}}$$

### Aufgabe 2

Differenzieren Sie folgende Wurzelfunktion:

$$f(x)=\sqrt{x^3-2x}$$

#### Rewrite Function
Rewrite function as  $f(x) = (x^3 - 2x)^{1/2}$.

Define $g(x) = x^{1/2}$, and $h(x) = x^3 - 2x$.

Consequently $f(x)$ can be re-written as = $f(x) = g(h(x))$

#### Find Derivative Using Chain Rule
$$\boxed{\frac{d}{dx}[g(h(x))] = g'(h(x)) \cdot h'(x)}$$

##### Sub-Derivatives
Derivative of g is $g'(x) = \frac{1}{2}x^-\frac{1}{2} = \frac{1}{2\sqrt{x}}$.
Derivative of h is $h'(x) = 3x^2 - 2$.

#### Plug-in Values

$$
  f'(x) = \frac{1}{2\sqrt{x^3 - 2x}} \cdot (3x^2 - 2)
        = \frac{3x^2 - 2}{2\sqrt{x^3 - 2x}}
$$



#### Check Calculations

```python
der = diff(sqrt(x**3 - 2*x), x)
der.simplify()
```

#### Therefore, the Solution Is :
$$\boxed{f'(x) = \frac{3x^2 - 2}{2\sqrt{x(x^2 - 2)}}}$$
