### Aufgabe 1a

Berechnen Sie $y'$ durch implizite Differentiation:

$$F(x,y) = x^2 + y^2 - r^2 = 0, \quad r \in \mathbb{R}$$

**Idee der impliziten Differentiation:** $y$ ist implizit als Funktion von $x$ definiert, d.h. $y = y(x)$. Wir leiten beide Seiten der Gleichung nach $x$ ab und wenden dabei die Kettenregel an.

**Schritt 1:** Ausgangsfunktion
$$x^2 + y^2 - r^2 = 0$$

**Schritt 2:** Beide Seiten nach $x$ ableiten. Da $r$ eine Konstante ist, verschwindet $r^2$. Der Term $y^2$ wird mit der Kettenregel abgeleitet: $\frac{d}{dx}[y^2] = 2y \cdot y'$.

$$\frac{d}{dx}\left[x^2\right] + \frac{d}{dx}\left[y^2\right] - \frac{d}{dx}\left[r^2\right] = 0$$

$$2x + 2y \cdot y' - 0 = 0$$

**Schritt 3:** Gleichung nach $y'$ auflösen:
$$2y \cdot y' = -2x$$

$$\boxed{y' = -\frac{x}{y}}$$

> **Geometrische Interpretation:** Dies ist die Kreisgleichung mit Radius $r$. Die Steigung $y' = -x/y$ beschreibt den Anstieg der Tangente an den Kreis im Punkt $(x,y)$. Der Radiusvektor $(x,y)$ steht senkrecht auf der Tangente, was durch das negative Kehrwertverhältnis bestätigt wird.

```python
x = symbols('x', real=True)
r_sym = symbols('r', real=True, positive=True)
y_fn = Function('y')(x)

F1a = x**2 + y_fn**2 - r_sym**2
print("F(x,y) =", F1a)

dF1a = F1a.diff(x)
print("dF/dx  =", dF1a)

sol1a = solveset(dF1a, y_fn.diff(x))
print("y'     =", sol1a)
```

### Aufgabe 1b

Berechnen Sie $y'$ durch implizite Differentiation:

$$e^{xy} = 2$$

**Schritt 1:** Ausgangsfunktion
$$e^{xy} = 2$$

**Schritt 2:** Beide Seiten nach $x$ ableiten. Die linke Seite erfordert die **Kettenregel** auf $e^{u}$ mit $u = x \cdot y(x)$, und anschließend die **Produktregel** für $u = x \cdot y(x)$:

$$\frac{d}{dx}\left[e^{xy}\right] = e^{xy} \cdot \frac{d}{dx}\left[x \cdot y\right] = e^{xy} \cdot \left(1 \cdot y + x \cdot y'\right)$$

Rechte Seite: $\frac{d}{dx}[2] = 0$

$$e^{xy}(y + x y') = 0$$

**Schritt 3:** Da $e^{xy} = 2 \neq 0$ für alle $(x,y)$, kann der Faktor $e^{xy}$ durch Division eliminiert werden:

$$y + x y' = 0$$

**Schritt 4:** Nach $y'$ auflösen:

$$x y' = -y$$

$$\boxed{y' = -\frac{y}{x}}$$

> **Bemerkung:** Aus $e^{xy} = 2$ folgt $xy = \ln 2$, also ist $y = \frac{\ln 2}{x}$ die explizite Lösung. Direkte Differentiation liefert $y' = -\frac{\ln 2}{x^2} = -\frac{y}{x}$ ✓

```python
x = symbols('x', real=True)
y_fn = Function('y')(x)

F1b = exp(x * y_fn) - 2
print("F(x,y) =", F1b)

dF1b = F1b.diff(x)
print("dF/dx  =", dF1b)

sol1b = solveset(dF1b, y_fn.diff(x))
print("y'     =", sol1b)
```

### Aufgabe 2a

Berechnen Sie $y'$ durch implizite Differentiation:

$$xy - \log y = 2$$

**Schritt 1:** Ausgangsfunktion (hier bezeichnet $\log$ den natürlichen Logarithmus $\ln$)
$$x \cdot y - \ln y = 2$$

**Schritt 2:** Beide Seiten nach $x$ ableiten:
- Term $x \cdot y$: **Produktregel** $\Rightarrow \frac{d}{dx}[x \cdot y] = 1 \cdot y + x \cdot y' = y + xy'$
- Term $\ln y$: **Kettenregel** $\Rightarrow \frac{d}{dx}[\ln y] = \frac{1}{y} \cdot y'$
- Rechte Seite: $\frac{d}{dx}[2] = 0$

$$y + x y' - \frac{y'}{y} = 0$$

**Schritt 3:** Alle Terme mit $y'$ auf eine Seite bringen:

$$xy' - \frac{y'}{y} = -y$$

**Schritt 4:** $y'$ ausklammern:

$$y' \left(x - \frac{1}{y}\right) = -y$$

**Schritt 5:** Den gemeinsamen Nenner $y$ bilden:

$$y' \cdot \frac{xy - 1}{y} = -y$$

**Schritt 6:** Nach $y'$ auflösen:

$$y' = -y \cdot \frac{y}{xy - 1}$$

$$\boxed{y' = -\frac{y^2}{xy - 1}}$$

```python
x = symbols('x', real=True)
y_fn = Function('y')(x)

F2a = x * y_fn - log(y_fn) - 2
print("F(x,y) =", F2a)

dF2a = F2a.diff(x)
print("dF/dx  =", dF2a)

sol2a = solveset(dF2a, y_fn.diff(x))
print("y'     =", sol2a)
```

Berechnen Sie $y'$ durch implizite Differentiation:

$$yx^2 - e^y = 0$$

**Schritt 1:** Ausgangsfunktion
$$y \cdot x^2 - e^y = 0$$

**Schritt 2:** Beide Seiten nach $x$ ableiten:

- Term $e^y$: **Kettenregel** $\Rightarrow \frac{d}{dx}[e^y] = e^y \cdot y'$

$$y' x^2 + 2xy - e^y y' = 0$$

**Schritt 3:** Terme mit $y'$ auf die linke Seite, Rest auf die rechte Seite:

$$y' x^2 - e^y y' = -2xy$$

**Schritt 4:** $y'$ ausklammern:

$$y'(x^2 - e^y) = -2xy$$

**Schritt 5:** Nach $y'$ auflösen:

$$y' = \frac{-2xy}{x^2 - e^y}$$

Zähler und Nenner mit $(-1)$ multiplizieren:

$$\boxed{y' = \frac{2xy}{e^y - x^2}}$$

```python
x = symbols('x', real=True)
y_fn = Function('y')(x)

F2b = y_fn * x**2 - exp(y_fn)
print("F(x,y) =", F2b)

dF2b = F2b.diff(x)
print("dF/dx  =", dF2b)

sol2b = solveset(dF2b, y_fn.diff(x))
print("y'     =", sol2b)
```

### Aufgabe 3

Ermitteln Sie $y'$ durch Logarithmieren und anschließendes Ableiten:

$$y = \sqrt[x]{x} = x^{1/x}$$

**Idee des logarithmischen Differenzierens:** Bei Funktionen der Form $f(x)^{g(x)}$ (Variable sowohl in Basis als auch im Exponenten) logarithmieren wir zuerst beide Seiten, um den Exponenten herunterzuholen.

**Schritt 1:** Logarithmieren beider Seiten:
$$\ln y = \ln\left(x^{1/x}\right) = \frac{1}{x} \cdot \ln x$$

**Schritt 2:** Beide Seiten nach $x$ ableiten. Die linke Seite ergibt mit der Kettenregel $\frac{y'}{y}$. Die rechte Seite wird mit der **Produktregel** differenziert mit $u = \frac{1}{x} = x^{-1}$ und $v = \ln x$:

$$\frac{d}{dx}\left[\frac{1}{x} \cdot \ln x\right] = \underbrace{\left(-\frac{1}{x^2}\right)}_{u'} \cdot \underbrace{\ln x}_{v} + \underbrace{\frac{1}{x}}_{u} \cdot \underbrace{\frac{1}{x}}_{v'}$$

$$= -\frac{\ln x}{x^2} + \frac{1}{x^2} = \frac{1 - \ln x}{x^2}$$

Also:
$$\frac{y'}{y} = \frac{1 - \ln x}{x^2}$$

**Schritt 3:** Mit $y = x^{1/x}$ multiplizieren:

$$\boxed{y' = x^{1/x} \cdot \frac{1 - \ln x}{x^2}}$$

> **Hinweis:** Das Vorzeichen von $y'$ hängt von $(1 - \ln x)$ ab. Bei $x = e$ gilt $\ln e = 1$, also $y'(e) = 0$ — dort liegt ein Extremum vor.

```python
x = symbols('x', positive=True)
y3 = x**(1/x)

# Direktes Differenzieren
dy3 = diff(y3, x)
print("y'  (roh)         =", dy3)
print("y'  (vereinfacht) =", dy3.simplify())

# Verifikation via logarithmisches Differenzieren
ln_y = log(y3).simplify()
print("\nln(y)          =", ln_y)
print("d/dx[ln(y)]    =", diff(ln_y, x).simplify())
print("y' = y * d/dx[ln(y)] =", (y3 * diff(ln_y, x)).simplify())
```

### Aufgabe 4

Ermitteln Sie $y'$ durch Logarithmieren und anschließendes Ableiten:

$$y = (\sin x)^{\cos x}$$

**Schritt 1:** Logarithmieren beider Seiten:
$$\ln y = \ln\left[(\sin x)^{\cos x}\right] = \cos x \cdot \ln(\sin x)$$

**Schritt 2:** Beide Seiten nach $x$ ableiten. Links ergibt die Kettenregel $\frac{y'}{y}$.

Rechts wird die **Produktregel** angewendet mit $u = \cos x$ und $v = \ln(\sin x)$:

$$u' = -\sin x, \qquad v' = \frac{1}{\sin x} \cdot \cos x = \frac{\cos x}{\sin x}$$

$$\frac{d}{dx}[\cos x \cdot \ln(\sin x)] = (-\sin x) \cdot \ln(\sin x) + \cos x \cdot \frac{\cos x}{\sin x}$$

$$= -\sin x \ln(\sin x) + \frac{\cos^2 x}{\sin x}$$

Also:
$$\frac{y'}{y} = \frac{\cos^2 x}{\sin x} - \sin x \ln(\sin x)$$

**Schritt 3:** Mit $y = (\sin x)^{\cos x}$ multiplizieren:

$$\boxed{y' = (\sin x)^{\cos x} \left(\frac{\cos^2 x}{\sin x} - \sin x \ln(\sin x)\right)}$$

Alternativ mit gemeinsamen Nenner $\sin x$:

$$y' = (\sin x)^{\cos x} \cdot \frac{\cos^2 x - \sin^2 x \ln(\sin x)}{\sin x}$$

```python
x = symbols('x', real=True)
y4 = sin(x)**cos(x)

dy4 = diff(y4, x)
print("y'  (roh)         =", dy4)
print("y'  (vereinfacht) =", dy4.simplify())

print("\nln(y)          =", log(y4).simplify())
print("d/dx[ln(y)]    =", diff(log(y4), x).simplify())
```

### Aufgabe 5

Ermitteln Sie $y'$ durch Logarithmieren und anschließendes Ableiten:

$$y = \sqrt[5]{\frac{(x-1)(x+2)}{x^2-3}}$$

**Vorteil des logarithmischen Differenzierens hier:** Produkte werden zu Summen, Quotienten zu Differenzen, und die 5. Wurzel wird zum Faktor $\frac{1}{5}$. Das ist deutlich einfacher als Quotienten- und Kettenregel direkt anzuwenden.

**Schritt 1:** Funktion als Potenz schreiben:
$$y = \left[\frac{(x-1)(x+2)}{x^2-3}\right]^{1/5}$$

**Schritt 2:** Logarithmieren beider Seiten und Logarithmusregeln anwenden:
$$\ln y = \frac{1}{5} \cdot \ln\left[\frac{(x-1)(x+2)}{x^2-3}\right]$$

$$\ln y = \frac{1}{5}\left[\ln(x-1) + \ln(x+2) - \ln(x^2-3)\right]$$

**Schritt 3:** Beide Seiten nach $x$ ableiten (jeder Term einzeln):

$$\frac{d}{dx}[\ln(x-1)] = \frac{1}{x-1}, \qquad \frac{d}{dx}[\ln(x+2)] = \frac{1}{x+2}, \qquad \frac{d}{dx}[\ln(x^2-3)] = \frac{2x}{x^2-3}$$

Also:
$$\frac{y'}{y} = \frac{1}{5}\left[\frac{1}{x-1} + \frac{1}{x+2} - \frac{2x}{x^2-3}\right]$$

**Schritt 4:** Mit $y$ multiplizieren:

$$\boxed{y' = \frac{1}{5} \cdot \sqrt[5]{\frac{(x-1)(x+2)}{x^2-3}} \cdot \left[\frac{1}{x-1} + \frac{1}{x+2} - \frac{2x}{x^2-3}\right]}$$

**Schritt 5 (optional):** Den Klammerausdruck auf gemeinsamen Nenner $(x-1)(x+2)(x^2-3)$ bringen:

$(x+2)(x^2-3) = x^3 + 2x^2 - 3x - 6$

$(x-1)(x^2-3) = x^3 - x^2 - 3x + 3$

$2x(x-1)(x+2) = 2x(x^2+x-2) = 2x^3 + 2x^2 - 4x$

Zähler der Klammer:
$$(x^3+2x^2-3x-6) + (x^3-x^2-3x+3) - (2x^3+2x^2-4x) = -x^2 - 2x - 3$$

```python
x = symbols('x', real=True)
y5 = ((x-1)*(x+2)/(x**2-3))**Rational(1, 5)

ln_y5 = log(y5).expand(force=True)
print("ln(y)            =", ln_y5)

dy5_ln = diff(ln_y5, x).simplify()
print("d/dx[ln(y)]      =", dy5_ln)

dy5 = diff(y5, x).simplify()
print("\ny' (vereinfacht) =", dy5)
print("y' (faktorisiert) =", dy5.factor())
```

### Aufgabe 6

Berechnen Sie im Punkt $P_0(1\,|\,y_0)$ die Krümmung der Kurve $y = x^3$.

$$\kappa = \frac{y''}{(1 + y'^2)^{3/2}}$$

**Schritt 1:** Erste und zweite Ableitung bestimmen:
$$y = x^3$$
$$y' = 3x^2$$
$$y'' = 6x$$

**Schritt 2:** Den Punkt $P_0$ bestimmen. Bei $x = 1$ gilt:
$$y_0 = 1^3 = 1, \quad \text{also } P_0(1\,|\,1)$$

**Schritt 3:** Ableitungen im Punkt $x = 1$ auswerten:
$$y'(1) = 3 \cdot 1^2 = 3$$
$$y''(1) = 6 \cdot 1 = 6$$

**Schritt 4:** In die Krümmungsformel einsetzen:
$$\kappa(1) = \frac{y''(1)}{\left(1 + [y'(1)]^2\right)^{3/2}} = \frac{6}{(1 + 9)^{3/2}} = \frac{6}{10^{3/2}}$$

**Schritt 5:** Vereinfachen:
$$10^{3/2} = 10 \cdot \sqrt{10}$$

$$\kappa(1) = \frac{6}{10\sqrt{10}} = \frac{3}{5\sqrt{10}}$$

Rationalisieren (Nenner rational machen):

$$\kappa(1) = \frac{3}{5\sqrt{10}} \cdot \frac{\sqrt{10}}{\sqrt{10}} = \frac{3\sqrt{10}}{50}$$

$$\boxed{\kappa(1) = \frac{3\sqrt{10}}{50} \approx 0{,}1897}$$

```python
x = symbols('x', real=True)

y6   = x**3
dy6  = diff(y6, x)
ddy6 = diff(y6, x, 2)

kappa6 = ddy6 / (1 + dy6**2)**Rational(3, 2)

print("y(x)  =", y6)
print("y'(x) =", dy6)
print("y''(x)=", ddy6)
print("\nKrümmungsformel κ(x) =", kappa6)

kappa_at1 = kappa6.subs(x, 1).simplify()
print("\nκ(1) =", kappa_at1)
print("κ(1) ≈", float(kappa_at1.evalf(6)))
```

### Aufgabe 7

Bestimmen Sie den Krümmungsradius und den Krümmungsmittelpunkt im Punkt $P_0(1\,|\,y_0)$ der Kurve $y = \sqrt{x}$.

$$\rho = \frac{1}{|\kappa|}, \qquad x_M = x - \frac{y'}{y''}(1+y'^2), \qquad y_M = y + \frac{1}{y''}(1+y'^2)$$

**Schritt 1:** Erste und zweite Ableitung bestimmen:
$$y = \sqrt{x} = x^{1/2}$$
$$y' = \frac{1}{2}x^{-1/2} = \frac{1}{2\sqrt{x}}$$
$$y'' = -\frac{1}{4}x^{-3/2} = -\frac{1}{4x^{3/2}}$$

**Schritt 2:** Punkt $P_0$ und Ableitungswerte bei $x = 1$:
$$y_0 = \sqrt{1} = 1, \quad P_0(1\,|\,1)$$
$$y'(1) = \frac{1}{2}, \qquad y''(1) = -\frac{1}{4}$$

**Schritt 3:** Hilfsgröße $1 + y'^2$ berechnen:
$$1 + [y'(1)]^2 = 1 + \frac{1}{4} = \frac{5}{4}$$

**Schritt 4:** Krümmung $\kappa$:
$$\kappa(1) = \frac{-1/4}{(5/4)^{3/2}} = \frac{-1/4}{\dfrac{5\sqrt{5}}{8}} = -\frac{1}{4} \cdot \frac{8}{5\sqrt{5}} = -\frac{2}{5\sqrt{5}} = -\frac{2\sqrt{5}}{25}$$

**Schritt 5:** Krümmungsradius:
$$\rho = \frac{1}{|\kappa|} = \frac{25}{2\sqrt{5}} = \frac{25\sqrt{5}}{10} = \frac{5\sqrt{5}}{2}$$

$$\boxed{\rho = \frac{5\sqrt{5}}{2} \approx 5{,}590}$$

**Schritt 6:** Krümmungsmittelpunkt $M(x_M\,|\,y_M)$:

$$x_M = 1 - \frac{y'(1)}{y''(1)} \cdot \left(1 + [y'(1)]^2\right) = 1 - \frac{1/2}{-1/4} \cdot \frac{5}{4} = 1 - (-2) \cdot \frac{5}{4} = 1 + \frac{10}{4} = \frac{7}{2}$$

$$y_M = y_0 + \frac{1}{y''(1)} \cdot \left(1 + [y'(1)]^2\right) = 1 + \frac{1}{-1/4} \cdot \frac{5}{4} = 1 + (-4) \cdot \frac{5}{4} = 1 - 5 = -4$$

$$\boxed{M\!\left(\frac{7}{2}\,\bigg|\,-4\right)}$$

> **Probe:** Der Abstand $|P_0 M|$ muss gleich $\rho$ sein:
> $$|P_0 M| = \sqrt{\left(\frac{7}{2}-1\right)^2 + (-4-1)^2} = \sqrt{\frac{25}{4} + 25} = \sqrt{\frac{125}{4}} = \frac{5\sqrt{5}}{2} = \rho \checkmark$$

```python
x = symbols('x', positive=True)

y7   = sqrt(x)
dy7  = diff(y7, x)
ddy7 = diff(y7, x, 2)

kappa7 = ddy7 / (1 + dy7**2)**Rational(3, 2)
rho7   = 1 / Abs(kappa7)
xM7    = x - (dy7 / ddy7) * (1 + dy7**2)
yM7    = y7 + (1 / ddy7) * (1 + dy7**2)

x0 = 1
print("--- Allgemeine Ausdrücke ---")
print("y'(x)  =", dy7)
print("y''(x) =", ddy7)
print("κ(x)   =", kappa7.simplify())
print("ρ(x)   =", rho7.simplify())
print()
print("--- Auswertung bei x = 1 ---")
print("y'(1)  =", dy7.subs(x, x0))
print("y''(1) =", ddy7.subs(x, x0))
print("κ(1)   =", kappa7.subs(x, x0).simplify(), "≈", float(kappa7.subs(x, x0).evalf(5)))
print("ρ(1)   =", rho7.subs(x, x0).simplify(), "≈", float(rho7.subs(x, x0).evalf(5)))
print("xM     =", xM7.subs(x, x0).simplify())
print("yM     =", yM7.subs(x, x0).simplify())

# Probe
xM_val = xM7.subs(x, x0)
yM_val = yM7.subs(x, x0)
dist = sqrt((xM_val - x0)**2 + (yM_val - y7.subs(x, x0))**2).simplify()
print("\nProbe |P0M| =", dist, "= ρ ?", dist == rho7.subs(x, x0).simplify())
```

```python
import numpy as np
import matplotlib.pyplot as plt

xM_num  = float(xM7.subs(x, x0))
yM_num  = float(yM7.subs(x, x0))
rho_num = float(rho7.subs(x, x0))

x_vals = np.linspace(0.01, 5, 500)
y_vals = np.sqrt(x_vals)

theta = np.linspace(0, 2 * np.pi, 500)
xK = xM_num + rho_num * np.cos(theta)
yK = yM_num + rho_num * np.sin(theta)

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x_vals, y_vals, 'b', lw=2.5, label=r'$y = \sqrt{x}$')
ax.plot(xK, yK, 'r--', lw=1.8,
        label=fr'Krümmungskreis ($\rho = 5\sqrt{{5}}/2 \approx {rho_num:.3f}$)')
ax.plot(x0, np.sqrt(x0), 'ko', ms=9, zorder=5, label=r'$P_0(1\,|\,1)$')
ax.plot(xM_num, yM_num, 'r^', ms=9, zorder=5,
        label=fr'$M(7/2\,|\,-4)$')
ax.plot([x0, xM_num], [np.sqrt(x0), yM_num], 'k:', lw=1.5, label=fr'Radius $\rho \approx {rho_num:.3f}$')

ax.set_xlim(-3, 7)
ax.set_ylim(-8, 5)
ax.set_aspect('equal')
ax.axhline(0, color='k', lw=0.7)
ax.axvline(0, color='k', lw=0.7)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
ax.set_title(r'Krümmungskreis von $y=\sqrt{x}$ im Punkt $P_0(1|1)$', fontsize=13)
plt.tight_layout()
plt.show()
```

### Aufgabe 8

Wo ist die Kurve $y = \log x$ am stärksten gekrümmt?

**Hinweis:** Bestimmen Sie $\kappa$, substituieren Sie $z = 1/x^2$ und finden Sie das Maximum für $\kappa^2$.

**Schritt 1:** Erste und zweite Ableitung bestimmen:
$$y = \ln x \quad (x > 0)$$
$$y' = \frac{1}{x}$$
$$y'' = -\frac{1}{x^2}$$

**Schritt 2:** Krümmung aufstellen und vereinfachen:
$$\kappa = \frac{y''}{(1+y'^2)^{3/2}} = \frac{-1/x^2}{\left(1 + 1/x^2\right)^{3/2}}$$

Zähler und Nenner mit $x^3$ erweitern:

$$\kappa = \frac{-x^3/x^2}{x^3 \cdot \left(\dfrac{x^2+1}{x^2}\right)^{3/2}} = \frac{-x}{(x^2+1)^{3/2}}$$

**Schritt 3:** $\kappa^2$ bilden (einfacher zu maximieren, da keine Absolutwerte nötig):
$$\kappa^2 = \frac{x^2}{(x^2+1)^3}$$

**Schritt 4:** Substitution $z = \dfrac{1}{x^2}$, also $x^2 = \dfrac{1}{z}$, $z > 0$:

- Zähler: $x^2 = \dfrac{1}{z}$

- Nenner: $(x^2 + 1)^3 = \left(\dfrac{1}{z} + 1\right)^3 = \left(\dfrac{1+z}{z}\right)^3 = \dfrac{(1+z)^3}{z^3}$

$$\kappa^2 = \frac{1/z}{(1+z)^3/z^3} = \frac{z^3}{z \cdot (1+z)^3} = \frac{z^2}{(1+z)^3}$$

**Schritt 5:** Maximum von $\kappa^2(z)$ durch Ableiten. Mit der Quotientenregel:

$$\frac{d(\kappa^2)}{dz} = \frac{2z \cdot (1+z)^3 - z^2 \cdot 3(1+z)^2}{(1+z)^6}$$

$(1+z)^2$ aus dem Zähler ausklammern:

$$= \frac{(1+z)^2 \left[2z(1+z) - 3z^2\right]}{(1+z)^6} = \frac{2z + 2z^2 - 3z^2}{(1+z)^4} = \frac{2z - z^2}{(1+z)^4} = \frac{z(2-z)}{(1+z)^4}$$

**Schritt 6:** Bedingung $\dfrac{d(\kappa^2)}{dz} = 0$ lösen:

$$z(2-z) = 0 \quad \Rightarrow \quad z = 0 \text{ (kein Max.)} \quad \text{oder} \quad z = 2 \checkmark$$

Vorzeichen-Wechsel: für $0 < z < 2$ ist $\frac{d(\kappa^2)}{dz} > 0$ (steigend), für $z > 2$ ist $\frac{d(\kappa^2)}{dz} < 0$ (fallend) → Maximum bei $z = 2$ ✓

**Schritt 7:** Rücksubstitution $z = \dfrac{1}{x^2} = 2$:

$$x^2 = \frac{1}{2} \quad \Rightarrow \quad x = \frac{1}{\sqrt{2}} = \frac{\sqrt{2}}{2}$$

$$\boxed{x^* = \frac{1}{\sqrt{2}} \approx 0{,}707}$$

Der zugehörige Kurvenpunkt ist:
$$P^* = \left(\frac{1}{\sqrt{2}}\,\bigg|\,\ln\frac{1}{\sqrt{2}}\right) = \left(\frac{1}{\sqrt{2}}\,\bigg|\,-\frac{\ln 2}{2}\right) \approx (0{,}707\,|\,-0{,}347)$$

```python
x = symbols('x', positive=True)
z = symbols('z', positive=True)

y8   = log(x)
dy8  = diff(y8, x)
ddy8 = diff(y8, x, 2)
kappa8 = ddy8 / (1 + dy8**2)**Rational(3, 2)

print("y'     =", dy8)
print("y''    =", ddy8)
print("κ(x)   =", kappa8.simplify())
print("κ²(x)  =", (kappa8**2).simplify())

# Nach Substitution z = 1/x^2
kappa_sq_z = z**2 / (1 + z)**3
print("\nNach z = 1/x²: κ²(z) =", kappa_sq_z)

# Ableitung
dkappa_sq_z = diff(kappa_sq_z, z).simplify()
print("d(κ²)/dz         =", dkappa_sq_z)
print("d(κ²)/dz (faktor)=", factor(dkappa_sq_z))

# Nullstellen
crit_z = solve(dkappa_sq_z, z)
print("\nKritische z-Werte:", crit_z)

# Rücksubstitution
z_star = 2
x_star = 1 / sqrt(z_star)
y_star = log(x_star)
print("\nz* =", z_star)
print("x* = 1/√2 =", x_star.simplify(), "≈", float(x_star.evalf(5)))
print("y* = ln(1/√2) =", y_star.simplify(), "≈", float(y_star.evalf(5)))
```
