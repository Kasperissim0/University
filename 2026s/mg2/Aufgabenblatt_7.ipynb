## Aufgabe 1

$$f(x)=x^2, \quad g(x)=\sqrt{x}$$

Gesucht: Fläche zwischen $f$ und $g$ im Intervall zwischen den reellwertigen Schnittpunkten.

### Analytische Lösung

#### Schritt 1 – Schnittpunkte bestimmen

Setze $f(x) = g(x)$:
$$x^2 = \sqrt{x} \implies x^4 = x \implies x^4 - x = 0 \implies x(x^3-1) = 0$$

$$\Rightarrow x_1 = 0, \quad x_2 = 1$$

Das Integrationsintervall ist $[0, 1]$.

#### Schritt 2 – Welche Funktion liegt oben?

Probe bei $x = \tfrac{1}{4}$:
$$f\!\left(\tfrac{1}{4}\right) = \tfrac{1}{16} = 0{,}0625, \qquad g\!\left(\tfrac{1}{4}\right) = \tfrac{1}{2} = 0{,}5$$

Also $g(x) \geq f(x)$ auf $[0,1]$.

#### Schritt 3 – Flächenintegral aufstellen und berechnen

$$A = \int_0^1 \bigl(g(x) - f(x)\bigr)\,dx = \int_0^1 \bigl(\sqrt{x} - x^2\bigr)\,dx$$

$$= \left[\frac{2}{3}x^{3/2} - \frac{x^3}{3}\right]_0^1 = \frac{2}{3} - \frac{1}{3} - 0$$

$$\boxed{A = \frac{1}{3}}$$


```python
# ── SymPy-Verifikation ─────────────────────────────────────────────────────────
f1 = x**2
g1 = sqrt(x)

# Schnittpunkte
schnitt1 = solve(f1 - g1, x)
print('Schnittpunkte:', schnitt1)

# Fläche
A1 = integrate(g1 - f1, (x, 0, 1))
print('Fläche A =', A1, '=', float(A1))

# ── Plot ──────────────────────────────────────────────────────────────────────
x_vals = np.linspace(0, 1.3, 500)
f_vals = x_vals**2
g_vals = np.sqrt(x_vals)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x_vals, f_vals, color=CYAN,   lw=2.5, label=r'$f(x) = x^2$')
ax.plot(x_vals, g_vals, color=ORANGE, lw=2.5, label=r'$g(x) = \sqrt{x}$')

x_fill = np.linspace(0, 1, 300)
ax.fill_between(x_fill, x_fill**2, np.sqrt(x_fill), alpha=0.25, color=GREEN,
                label=r'Fläche $A = \frac{1}{3}$')

ax.axhline(0, color='#444', lw=0.5)
ax.axvline(0, color='#444', lw=0.5)
ax.scatter([0, 1], [0, 1], color='white', zorder=5, s=60, label='Schnittpunkte')
ax.set(xlim=(-0.1, 1.3), ylim=(-0.05, 1.3),
       xlabel='x', ylabel='y',
       title=r'Aufgabe 1 – Fläche zwischen $f(x)=x^2$ und $g(x)=\sqrt{x}$')
ax.grid(True)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## Aufgabe 2

$$f(x) = 4x, \quad g(x) = x^2 + 2x - 3$$

Gesucht: Fläche zwischen $f$ und $g$ im Intervall zwischen den reellwertigen Schnittpunkten.

### Analytische Lösung

#### Schritt 1 – Schnittpunkte bestimmen

Setze $f(x) = g(x)$:
$$4x = x^2 + 2x - 3 \implies 0 = x^2 - 2x - 3 = (x-3)(x+1)$$

$$\Rightarrow x_1 = -1, \quad x_2 = 3$$

Das Integrationsintervall ist $[-1, 3]$.

#### Schritt 2 – Welche Funktion liegt oben?

Probe bei $x = 0$:
$$f(0) = 0, \qquad g(0) = 0 + 0 - 3 = -3$$

Also $f(x) \geq g(x)$ auf $[-1, 3]$.

#### Schritt 3 – Flächenintegral aufstellen

$$A = \int_{-1}^{3} \bigl(f(x) - g(x)\bigr)\,dx = \int_{-1}^{3} \bigl(4x - x^2 - 2x + 3\bigr)\,dx = \int_{-1}^{3} \bigl(-x^2 + 2x + 3\bigr)\,dx$$

#### Schritt 4 – Integral berechnen

$$A = \left[-\frac{x^3}{3} + x^2 + 3x\right]_{-1}^{3}$$

**Obere Grenze** $x=3$:
$$-\frac{27}{3} + 9 + 9 = -9 + 18 = 9$$

**Untere Grenze** $x=-1$:
$$-\frac{-1}{3} + 1 - 3 = \frac{1}{3} - 2 = -\frac{5}{3}$$

$$A = 9 - \left(-\frac{5}{3}\right) = 9 + \frac{5}{3} = \frac{27 + 5}{3}$$

$$\boxed{A = \frac{32}{3} \approx 10{,}67}$$


```python
# ── SymPy-Verifikation ─────────────────────────────────────────────────────────
x2 = symbols('x', real=True)
f2 = 4*x2
g2 = x2**2 + 2*x2 - 3

schnitt2 = solve(f2 - g2, x2)
print('Schnittpunkte:', schnitt2)

A2 = integrate(f2 - g2, (x2, -1, 3))
print('Fläche A =', A2, '=', float(A2))

# ── Plot ──────────────────────────────────────────────────────────────────────
x_vals = np.linspace(-2, 4, 600)
f_vals = 4*x_vals
g_vals = x_vals**2 + 2*x_vals - 3

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x_vals, f_vals, color=CYAN,   lw=2.5, label=r'$f(x) = 4x$')
ax.plot(x_vals, g_vals, color=ORANGE, lw=2.5, label=r'$g(x) = x^2+2x-3$')

x_fill = np.linspace(-1, 3, 400)
ax.fill_between(x_fill, 4*x_fill, x_fill**2 + 2*x_fill - 3,
                alpha=0.25, color=GREEN, label=r'Fläche $A = \frac{32}{3}$')

ax.scatter([-1, 3], [-4, 12], color='white', zorder=5, s=60, label='Schnittpunkte')
ax.axhline(0, color='#444', lw=0.5)
ax.axvline(0, color='#444', lw=0.5)
ax.set(xlim=(-2, 4), ylim=(-8, 15),
       xlabel='x', ylabel='y',
       title=r'Aufgabe 2 – Fläche zwischen $f(x)=4x$ und $g(x)=x^2+2x-3$')
ax.grid(True)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

## Aufgabe 3

$$f(x) = \log x, \quad g(x) = (\log x)^2$$

*(Hier bezeichnet $\log$ den natürlichen Logarithmus $\ln$.)*

Gesucht: Fläche zwischen $f$ und $g$ im Intervall zwischen den reellwertigen Schnittpunkten.

### Analytische Lösung

#### Schritt 1 – Schnittpunkte bestimmen

Definitionsbereich: $x > 0$.

Setze $f(x) = g(x)$:
$$\ln x = (\ln x)^2 \implies \ln x \cdot (\ln x - 1) = 0$$

$$\Rightarrow \ln x = 0 \;\Rightarrow\; x_1 = 1, \qquad \ln x = 1 \;\Rightarrow\; x_2 = e$$

Das Integrationsintervall ist $[1, e]$.

#### Schritt 2 – Welche Funktion liegt oben?

Probe bei $x = e^{1/2} = \sqrt{e} \approx 1{,}65$:
$$f(\sqrt{e}) = \tfrac{1}{2}, \qquad g(\sqrt{e}) = \tfrac{1}{4}$$

Also $f(x) \geq g(x)$ auf $[1, e]$.

#### Schritt 3 – Flächenintegral aufstellen

$$A = \int_1^e \bigl(\ln x - (\ln x)^2\bigr)\,dx$$

#### Schritt 4 – Teilintegrale berechnen

**Integral 1:** $\displaystyle\int_1^e \ln x\,dx$

Mit partieller Integration ($u = \ln x$, $dv = dx$):
$$\int \ln x\,dx = x\ln x - x + C$$

$$\left[x\ln x - x\right]_1^e = (e \cdot 1 - e) - (1 \cdot 0 - 1) = 0 + 1 = 1$$

**Integral 2:** $\displaystyle\int_1^e (\ln x)^2\,dx$

Mit partieller Integration ($u = (\ln x)^2$, $dv = dx$):
$$\int (\ln x)^2\,dx = x(\ln x)^2 - 2\int \ln x\,dx = x(\ln x)^2 - 2x\ln x + 2x + C$$

$$\left[x(\ln x)^2 - 2x\ln x + 2x\right]_1^e = (e \cdot 1 - 2e + 2e) - (0 - 0 + 2) = e - 2$$

#### Schritt 5 – Zusammensetzen

$$A = 1 - (e-2) = 3 - e$$

$$\boxed{A = 3 - e \approx 0{,}282}$$


```python
# ── SymPy-Verifikation ─────────────────────────────────────────────────────────
x3 = symbols('x', real=True, positive=True)
f3 = log(x3)
g3 = log(x3)**2

schnitt3 = solve(f3 - g3, x3)
print('Schnittpunkte:', schnitt3)

A3 = integrate(f3 - g3, (x3, 1, E))
print('Fläche A =', simplify(A3), '=', float(A3))
print('Kontrolle: 3 - e =', 3 - float(E))

# ── Plot ──────────────────────────────────────────────────────────────────────
x_vals = np.linspace(0.1, 3.5, 600)
f_vals = np.log(x_vals)
g_vals = np.log(x_vals)**2

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x_vals, f_vals, color=CYAN,   lw=2.5, label=r'$f(x) = \ln x$')
ax.plot(x_vals, g_vals, color=ORANGE, lw=2.5, label=r'$g(x) = (\ln x)^2$')

x_fill = np.linspace(1, np.e, 400)
ax.fill_between(x_fill, np.log(x_fill), np.log(x_fill)**2,
                alpha=0.25, color=GREEN, label=r'Fläche $A = 3-e$')

ax.scatter([1, np.e], [0, 1], color='white', zorder=5, s=60, label='Schnittpunkte $(1,0)$ und $(e,1)$')
ax.axhline(0, color='#444', lw=0.5)
ax.axvline(0, color='#444', lw=0.5)
ax.set(xlim=(0.1, 3.5), ylim=(-1.5, 2.5),
       xlabel='x', ylabel='y',
       title=r'Aufgabe 3 – Fläche zwischen $\ln x$ und $(\ln x)^2$')
ax.grid(True)
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
